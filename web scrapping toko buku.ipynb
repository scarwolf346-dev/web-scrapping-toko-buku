```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from fake_useragent import UserAgent
import time

base_url = "http://books.toscrape.com/catalogue/page-{}.html"
ua = UserAgent()

book_data = []

# Scrape 5 halaman pertama (bisa ditambah)
for page in range(1, 6):
    print(f"Scraping page {page}...")
    headers = {'User-Agent': ua.random}
    url = base_url.format(page)

    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        print(f"Gagal mengambil halaman {page}")
        continue

    soup = BeautifulSoup(response.text, 'html.parser')
    books = soup.select('.product_pod')

    for book in books:
        title = book.h3.a['title']
        price = book.select_one('.price_color').text.strip()
        rating = book.select_one('.star-rating')['class'][1]  # e.g., 'Three'

        book_data.append({
            'Judul': title,
            'Harga': price,
            'Rating': rating
        })

    time.sleep(1)  # Delay agar tidak dianggap bot

# Simpan ke CSV
df = pd.DataFrame(book_data)
df.to_csv('daftar_buku.csv', index=False, encoding='utf-8-sig')
print("Selesai! Data disimpan di 'daftar_buku.csv'")
```
